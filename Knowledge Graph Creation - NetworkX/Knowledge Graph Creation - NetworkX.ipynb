```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create a new graph
G = nx.Graph()

# add nodes with types
G.add_node('Alice', type='Person')
G.add_node('Bob', type='Person')
G.add_node('Lucy', type='Person')
G.add_node('Natsu', type='Person')
G.add_node('Python', type='Programming Language')
G.add_node('AI', type='Concept')
```

```python
# add egdes/relationships for the nodes
G.add_edge('Alice', 'Bob', relationship='Friend')
G.add_edge('Lucy', 'Natsu', relationship='Friend')
G.add_edge('Alice', 'Lucy', relationship='Friend')
G.add_edge('Alice', 'Python', relationship='Knows')
G.add_edge('Python', 'AI', relationship='Implemented with')
G.add_edge('Bob', 'AI', relationship='Uses')
```

```python
# Draw the Graph
pos = nx.spring_layout(G)
plt.figure(figsize=(9, 7))

# draw nodes with colors based on type
node_types = [G.nodes[node]['type'] for node in G.nodes]
color_map = {'Person': 'lightblue', 'Programming Language': 'lightgreen', 'Concept': 'lightcoral'}
colors = [color_map[node_type] for node_type in node_types]

nx.draw(G, pos, with_labels=True, node_color=colors, node_size=2000, font_size=10, font_color='black', edge_color='gray')
edge_labels = nx.get_edge_attributes(G, 'relationship')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='blue')

# show the plot
plt.title('Knowledge Graph')
plt.show()
```
